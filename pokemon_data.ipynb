```python
pokemon["Total"] = (
    pokemon["HP"] + pokemon["Attack"] + pokemon["Defense"] +
    pokemon["Sp. Atk"] + pokemon["Sp. Def"] + pokemon["Speed"]
)

pokemon.nlargest(10, "Total")[["Dex_No", "Type 1","Type 2" ,"Total"]]
```

```python
pokemon["Attack_Level"] = pd.cut(
    pokemon["Attack"],
    bins=[0,50,100,150,200],
    labels=["Weak","Average","Strong","Elite"]
)

pokemon["Attack_Level"].value_counts()
```

```python
pokemon.plot.scatter(x="Attack", y="Defense")
plt.show()
```
